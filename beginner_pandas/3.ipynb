```python
countries = ['United States', 'India', 'United Kingdom', 'Germany', 'Canada']
filt = (df['Country'].isin(countries)) & (df['ConvertedComp'] > 70000)
df.loc[filt, ['Country', 'ConvertedComp']]
```

```python
filt = df['LanguageWorkedWith'].str.contains('Python', na=False) # display only people who worked with Python
df.loc[filt, 'LanguageWorkedWith']
```

```python
df.rename(columns={'ConvertedComp' : 'SalaryUSD'}, inplace=True)
df['SalaryUSD']
```

```python
# df['Country'].sort_values()

# sort Country in ascending order and Salary in descending 
df.sort_values(by=['Country','SalaryUSD'], inplace=True, ascending=[True, False])
df[['Country', 'SalaryUSD']].head(50)
```

```python
# Largest and smallest values -> return pandas.core.series
df['SalaryUSD'].nlargest(10)
df['SalaryUSD'].nsmallest(10)

# returns dataframe
df.nlargest(10, 'SalaryUSD')
```
