```python
# get countries from special country

country_group = df.groupby(['Country'])
```

```python
country_group['SocialMedia'].value_counts(normalize=True).loc['Russian Federation']
```

```python
country_group['SalaryUSD'].median().loc['Russian Federation']
```

```python
# aggregate function
country_group['SalaryUSD'].agg(['median', 'mean']).loc['Canada']
```

```python
# using lambda functions in series.groupby object
# country_group['LanguageWorkedWith'].apply(lambda x: x.str.contains('Python').value_counts())

# calculate the number of people who knows python in each country and add the new column to the data frame
from matplotlib import pyplot as plt
import numpy as np

country_respondents = df['Country'].value_counts()
know_python = country_group['LanguageWorkedWith'].apply(lambda x : x.str.contains('Python').sum())

python_df = pd.concat([country_respondents, know_python], axis='columns',sort=False)
python_df.rename(columns={'Country' : 'Num_of_responses', 'LanguageWorkedWith' : 'Know_Python'}, inplace=True)
python_df['PctKnowsPython'] = (python_df['Know_Python']/python_df['Num_of_responses']) * 100
python_df.sort_values(by='PctKnowsPython', ascending=False, inplace=True)


x_ticks = np.arange(len(python_df['PctKnowsPython'].head(25)))
y_ticks = python_df['PctKnowsPython'].head(25).to_list()
labels = python_df['PctKnowsPython'].head(25).index


plt.title("Percentage of people in countries who knows Python")
plt.barh(labels, y_ticks)
plt.tight_layout()
plt.show()



# python_df['PctKnowsPython'].loc['Russian Federation'] # pct of people in russian federation who knows python
```
