```python
import pandas as pd
from matplotlib import pyplot as plt

plt.style.use('seaborn-v0_8-dark')
ages = [18, 19, 21, 25, 26, 26, 30, 32, 38, 45, 55]
bins = [10, 20, 30, 40, 50, 60] #x axes

# bin can take integer or own data
plt.hist(ages, bins=bins, edgecolor='black')

plt.title('Ages of Respondents')
plt.xlabel('Ages')
plt.ylabel('Total Respondents')

plt.tight_layout()
```

```python
import numpy as np

plt.style.use('fivethirtyeight')

data = pd.read_csv('hists_7.csv')
ids  = data['Responder_id']
ages = data['Age']

median_age = data['Age'].median()

bins = [x for x in range(10, 101, 10)]

plt.hist(ages, bins=bins, edgecolor='black')
plt.axvline(median_age, color='red', label='median_age') # plot a median age line


plt.legend()

plt.title('Ages of respondents')
plt.ylabel('total respondents')
plt.xlabel('age of respondents')
plt.show()
```
