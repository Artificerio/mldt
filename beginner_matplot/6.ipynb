```python
import pandas as pd
from matplotlib import pyplot as plt
data = pd.read_csv('new_data.csv')
ages = data['Age']
dev_salaries = data['All_Devs']
py_salaries = data['Python']
js_salaries = data['JavaScript']


plt.plot(ages, dev_salaries, color='#4444', linestyle='--',
        label='all devs')
plt.plot(ages, py_salaries, color='c', label='python')

plt.title('Median Salary (USD) by Age')
plt.xlabel('age')
plt.ylabel('salary')

# y2 = plot down to the given threshold
# plot where the conditional is met
# interpolate clip the x intersections
# label is label...
overall_median = data['All_Devs'].mean()
plt.fill_between(ages, py_salaries, y2=dev_salaries, alpha=0.25,
    where=py_salaries > dev_salaries, interpolate=True,label='Above average')
    
plt.fill_between(ages, py_salaries, y2=dev_salaries, alpha=0.25,
    where=py_salaries <= dev_salaries, interpolate=True, label='Below average')

plt.legend(loc='upper left')
# plt.tight_layout()
plt.show()
data.describe()
```
